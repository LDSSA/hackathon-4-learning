# Intro to Recommender Systems

## 1 Framework

The general framework we are going to use includes three fundamental components:

* **Users** - the people in our system
* **Items** - the things we want to recommend to our users
* **Data** - the different ways an user can express opinions about our items, i.e. where users and items *meet*.

Then, there are three optional components, applied in specific models:

* **User attributes** - ways to *describe* our users
* **Item attributes** - ways to *describe* our items
* **User model** - a way to represent user preferences.
